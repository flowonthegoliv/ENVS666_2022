```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
#read in the data 
s1 = pd.read_csv('f008d1cc1110.csv',sep = ';')
s2 = pd.read_csv('f008d1cbdea4.csv',sep = ';')
times1 = pd.DatetimeIndex(s1['Date & Time'])
times2 = pd.DatetimeIndex(s2['Date & Time'])
pm2p5s1 = s1['PM2.5']
pm2p5s2 = s2['PM2.5']
```

```python
date_1 = '02/01/2022'

sub_s1 = pm2p5s1[times1 > date_1]
sub_t1 = times1[times1 > date_1]

sub_s2 = pm2p5s2[times2 > date_1]
sub_t2 = times2[times2 > date_1]

plt.plot(sub_t1,sub_s1)
plt.plot(sub_t2,sub_s2)

av_s1 = np.nanmean(sub_s1)
av_s2 = np.nanmean(sub_s2)

std_s1 = np.nanstd(sub_s1)
std_s2 = np.nanstd(sub_s2)

print(av_s1,av_s2)
print(std_s1,std_s2)
```

```python
plt.hist(sub_s1,bins=100,range=[0,20])
plt.hist(sub_s2,bins=100,range=[0,20])
```
